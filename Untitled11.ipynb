```python
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
```

```python
df['Ship Mode'].unique()
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
#deriving new columns-discount column-
df['discount']=df['list_price']*df['discount_percent']*0.01
```

```python
df.discount
```

```python
#making new column - sale price-
df['sale_price']= df['list_price']-df['discount']
```

```python
df
```

```python
#making a new column for profit
df['profit']=df['sale_price']-df['cost_price']
```

```python
df
```

```python
#converting order date from object to datetime
df['order_date']=pd.to_datetime(df['order_date'], format= "%Y-%m-%d")
```

```python
df.dtypes
```

```python
#dropping columns
df.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
### 1. Total Sales and Profit by Category
category_data = df.groupby('category').agg({'sale_price': 'sum', 'profit': 'sum'}).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='sale_price', data=category_data, color='skyblue', label='Sales')
sns.barplot(x='category', y='profit', data=category_data, color='orange', label='Profit')
plt.title('Total Sales and Profit by Category')
plt.ylabel('Amount')
plt.xlabel('Category')
plt.legend()
plt.show()
```

```python
### 2. Discount vs. Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discount', y='profit', hue='category', palette='viridis')
plt.title('Discount vs. Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
### 3. Sales Trend Over Time (Monthly)
df['month'] = df['order_date'].dt.to_period('M').astype(str)
monthly_sales = df.groupby('month').agg({'sale_price': 'sum'}).reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='sale_price', marker='o', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
